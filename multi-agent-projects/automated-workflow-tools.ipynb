````python
def add_objects(obj1, obj2):
   try:
       result = obj1 + obj2
       return result
   except TypeError:
       return "Error: Unable to add objects"

# Example usage
obj1 = 5
obj2 = 3
result = add_objects(obj1, obj2)
print(result)  # Output: 8
````

````python
# Here is the implementation of the function in Python:

def intersect(*lists):
    if not lists:
        return []
    
    intersection = set(lists[0])
    for lst in lists[1:]:
        intersection = intersection.intersection(lst)
    
    return sorted(list(intersection))

# Example
list1 = [1, 2, 3, 4]
list2 = [2, 3, 4, 5]
list3 = [3, 4, 5, 6]
result = intersect(list1, list2, list3)
print(result)  # Output: [3, 4]
````
